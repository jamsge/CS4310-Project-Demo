```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = "data1.csv"
df = pd.read_csv(file_path)

df["CompressionTime(ms)"] = df["CompressionTime(ns)"] / 1_000_000

df = df[df["Run"] > 2]
plt.figure(figsize=(10, 6))

for algo in df["Algorithm"].unique():
    subset = df[df["Algorithm"] == algo]
    plt.plot(subset["Run"], subset["CompressionTime(ms)"], label=algo)

plt.title("Compression Time per Run")
plt.xlabel("Run")
plt.ylabel("Compression Time (ms)")
plt.legend(title="Algorithm")
plt.grid(True)
plt.tight_layout()

plt.show()

df["DecompressionTime(ms)"] = df["DecompressionTime(ns)"] / 1_000_000

plt.figure(figsize=(10, 6))

for algo in df["Algorithm"].unique():
    subset = df[df["Algorithm"] == algo]
    plt.plot(subset["Run"], subset["DecompressionTime(ms)"], label=algo)

plt.title("Decompression Time per Run")
plt.xlabel("Run")
plt.ylabel("Decompression Time (ms)")
plt.legend(title="Algorithm")
plt.grid(True)
plt.tight_layout()

plt.show()
```
